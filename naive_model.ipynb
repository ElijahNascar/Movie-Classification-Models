```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
sns.set(font_scale=1.35)

movies = pd.read_csv('rotten_tomatoes_movies.csv')
reviews = pd.read_csv('rotten_tomatoes_movie_reviews.csv')
```

```python
# Averaging Reviews
reviews['bin_review'] = reviews['reviewState'].apply(lambda x: 1 if x == 'fresh' else 0)
avg_reviews = reviews.groupby('id').mean()

avg_reviews['reviewState'] = avg_reviews['bin_review'].apply(lambda x: 1 if x >= 0.5 else 0)

avg_reviews = avg_reviews.drop(columns=['reviewId', 'bin_review'])
avg_reviews.head()
```

```python
# Add Reviews to Movie Dataset

rotten_t = pd.merge(left=movies, right=avg_reviews, on='id')
rotten_t_naive = rotten_t.loc[:, ['audienceScore', 'runtimeMinutes', 'isTopCritic', 'reviewState']]

#Tried to make the categoricals work, computer broke cuz it doesnt have the memory
#rotten_t = rotten_t.iloc[0:34630]

#predictors = rotten_t.columns.tolist()
#predictors.remove('reviewState')
#predictors_categorical = predictors
#predictors_categorical.remove('audienceScore')
#predictors_categorical.remove('tomatoMeter')
#predictors_categorical.remove('runtimeMinutes')
#predictors_categorical.remove('isTopCritic')

#rotten_t_dummies = pd.get_dummies(rotten_t, columns=predictors_categorical)
#rotten_t_dummies.head()
```

```python
rotten_t_naive.dropna(inplace=True)
rotten_t_naive
```

```python
# Create Naive Model
y = rotten_t_naive.reviewState
X = rotten_t_naive.drop("reviewState", axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 45)

# Create a model object - not trained yet
logreg = LogisticRegression()

# Train the model
logreg.fit(X_train, y_train)
```

```python
# Get the predicted classes first
y_pred = logreg.predict(X_test)

# Use the predicted and true classes for accuracy
print(accuracy_score(y_pred, y_test)*100) 
```
